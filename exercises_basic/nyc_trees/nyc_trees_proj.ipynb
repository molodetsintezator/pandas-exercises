Обогащение данных


Создать новый столбец tree_category, где:
"Small" для tree_dbh < 5
"Medium" для tree_dbh от 5 до 20
"Large" для tree_dbh > 20
Посчитать, сколько деревьев попадает в каждую категорию.


```python
sizes = [
    (nyc_trees['tree_dbh'] < 5),
    (nyc_trees['tree_dbh'] >= 5)&(nyc_trees['tree_dbh'] <= 20),
    (nyc_trees['tree_dbh'] > 20)]
categories = ['Small', 'Medium', 'Large']
nyc_trees['tree_category'] = np.select(sizes, categories)

nyc_trees.groupby('tree_category', as_index = False).size()
```
